```python
import matplotlib.pyplot as plt
import numpy as np
res = 100
values = [int(np.sin(i)*100) for i in range(90)]
plt.xlim(-res, res)
plt.ylim(0, res+5)

def printline(angle,linelen):
    servoangle = np.pi*(angle/180)
    plt.plot([i*np.cos(servoangle) for i in range(linelen)],[i*np.sin(servoangle) for i in range(linelen)],'k')
def printvals(values):
    for i in range(45,135):
        printline(i,values[i-45])
printvals(values)
```
